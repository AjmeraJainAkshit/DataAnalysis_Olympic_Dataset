## Hypothesis Testing 

## 1) if gdp of country is best in position , then it win more medals

```python
medals_per_country = pd.crosstab(index = olympic_merge.Country, columns = olympic_merge.Medal,margins = True, margins_name = "Total")
medals_per_country
```

```python
medals_per_country.drop("Total", axis = 0, inplace = True)
```

```python
medals_per_country
```

```python
medals_per_country = medals_per_country[["Gold","Silver","Bronze","Total"]].sort_values(by = ["Gold","Silver","Bronze","Total"], ascending = False)
```

```python
medals_per_country
```

```python
medals_per_country.head(30)
```

```python
medals_per_country = medals_per_country.merge(dicti, how = "left", left_index = True, right_on = "Country").drop(columns = ["Code"]).set_index("Country")
```
